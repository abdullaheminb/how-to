```python
from langchain_openai import ChatOpenAI
from langchain_community.document_loaders import UnstructuredHTMLLoader
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.text_splitter import HTMLHeaderTextSplitter
from langchain_core.prompts import ChatPromptTemplate
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain.chains import create_retrieval_chain
from langchain.text_splitter import (
    Language,
    RecursiveCharacterTextSplitter,
)
```

```python
llm = ChatOpenAI(temperature=0, model="gpt-4")
loader = UnstructuredHTMLLoader("./docs.html")
data = loader.load()
```

```python
html_splitter = RecursiveCharacterTextSplitter.from_language(
    language=Language.HTML, chunk_size=60, chunk_overlap=0
)
documents = html_splitter.split_documents(data)
embeddings = OpenAIEmbeddings()
vector = FAISS.from_documents(documents, embeddings)
```

```python
prompt = ChatPromptTemplate.from_template("""read through the html, gain insights, and answer questions like a proper technical writer :
<context>{context}</context>
Question: {input}""")
document_chain = create_stuff_documents_chain(llm, prompt)
```

```python
retriever = vector.as_retriever()
retrieval_chain = create_retrieval_chain(retriever, document_chain)
```

```python
response = retrieval_chain.invoke({"input": "Can you compare minds db with an alternative"})
print(response["answer"])
```
